```python
import pytrends
from pytrends.request import TrendReq
from pytrends import dailydata
pytrend = TrendReq()
import time

import pandas as pd
```

```python
exact_keywords = ['Bitcoin', 'GeForce']
df_list_1 = []
for word in exact_keywords:
    thisword = dailydata.get_daily_data(word = word, start_year = 2022, start_mon = 2, stop_year = 2022, stop_mon = 4)
    thisword = thisword[thisword['isPartial'] == False]
    df_list_1.append(thisword[word])
    print(word + ' done\n')
    time.sleep(10)
```

```python
df1 = pd.concat(df_list_1, axis = 1)
df1.tail(30)
```

```python
exact_keywords = ['Bitcoin', 'GeForce']
df_list_2 = []
for word in exact_keywords:
    thisword = dailydata.get_daily_data(word = word, start_year = 2022, start_mon = 4, stop_year = 2022, stop_mon = 5)
    thisword = thisword[thisword['isPartial'] == False]
    df_list_2.append(thisword[word])
    print(word + ' done\n')
    time.sleep(10)
```

```python
df2 = pd.concat(df_list_2, axis = 1)
df2.head(30)
```

```python
bitcoin_1 = df1.tail(30)['Bitcoin']
bitcoin_2 = df2.head(30)['Bitcoin']
```

```python
geforce_1 = df1.tail(30)['GeForce']
geforce_2 = df2.head(30)['GeForce']
```

```python
bitcoin_1 / bitcoin_2
```

```python
geforce_1 / geforce_2
```
