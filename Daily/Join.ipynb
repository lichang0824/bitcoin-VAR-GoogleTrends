```python
import pandas as pd
```

## Clean trends

```python
trends = pd.read_csv('trends.csv')
```

```python
trends['date'] = pd.to_datetime(trends['date']).dt.date
```

```python
trends.set_index('date', inplace = True)
```

```python
trends
```

## Clean price

```python
price = pd.read_csv('price.csv')
```

```python
price['Date'] = pd.to_datetime(price['Date']).dt.date
price.rename(columns = {'Date' : 'date'}, inplace = True)
price.set_index('date', inplace = True)
```

```python
price = price['High'].to_frame().rename(columns = {'High' : 'Price'})
```

```python
price
```

## Join

```python
df = pd.concat([trends, price], axis = 1, join = 'inner')
df
```

```python
df.to_csv('data.csv')
```
