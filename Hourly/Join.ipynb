```python
import pandas as pd
```

## Clean trends

```python
trends = pd.read_csv('trends.csv')
trends['time'] = pd.to_datetime(trends['time'])
trends.set_index('time', inplace = True)
trends = trends[~trends.index.duplicated(keep='first')]
```

```python
trends
```

## Clean price

```python
price = pd.read_csv('price.csv')
```

```python
price['time'] = pd.to_datetime(price['time']).dt.tz_localize(None)
price.set_index('time', inplace = True)
```

```python
price
```

## Join

```python
df = pd.concat([trends, price['High'].to_frame().rename(columns = {'High' : 'Price'})], axis = 1, join = 'inner')
df
```

```python
df.to_csv('data.csv')
```
